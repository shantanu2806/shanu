```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

with open("C:\\Users\\Shubham Dhanawade\\OneDrive\\Desktop\\Downloads\\archive\\spotify-2023.csv", 'rb') as f:
    result = chardet.detect(f.read())


# Use the detected encoding
detected_encoding = result['encoding']
data = pd.read_csv("C:\\Users\\Shubham Dhanawade\\OneDrive\\Desktop\\Downloads\\archive\\spotify-2023.csv", encoding=detected_encoding)

# Sort the data by release date
data['released_year'] = pd.to_datetime(data['released_year'], errors='coerce')
data = data.dropna(subset=['released_year'])
data = data.sort_values(by='released_year')

# Create a figure and axis for the bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Function to update the bar chart for each frame of the animation
def update(frame):
    ax.clear()
    # Filter the data for a specific time frame (e.g., every 100th record)
    frame_data = data.iloc[frame * 100 : (frame + 1) * 100]
    # Create a bar chart of song popularity for this time frame
    ax.barh(frame_data['artist(s)_name'], frame_data['in_spotify_playlists'], color='skyblue')
    ax.set_xlabel('Popularity')
    ax.set_ylabel('Song Name')
    ax.set_title(f'Spotify Song Popularity Over Time (Frame {frame + 1})')

# Create the animation
num_frames = len(data) // 100  # Number of frames (adjust as needed)
animation = FuncAnimation(fig, update, frames=num_frames, repeat=False)

# Display the animation in a Jupyter Notebook or IPython environment
HTML(animation.to_jshtml())

```

```python
import pandas as pd

# Detect the encoding of the file
with open("C:\\Users\\Shubham Dhanawade\\OneDrive\\Desktop\\Downloads\\archive\\spotify-2023.csv", 'rb') as f:
    result = chardet.detect(f.read())


# Use the detected encoding
detected_encoding = result['encoding']
data = pd.read_csv("C:\\Users\\Shubham Dhanawade\\OneDrive\\Desktop\\Downloads\\archive\\spotify-2023.csv", encoding=detected_encoding)
# Data Cleaning Steps

df = pd.DataFrame(data)
#combine 3 columns
df['release_date'] = df['released_year'] + df['released_month'] + df['released_day']


# 4. Standardizing Formats
# Example: Convert release_date column to datetime format
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# 5. Dealing with Outliers
# Example: Remove rows with popularity scores outside a reasonable range
df = df[(df['in_spotify_playlists'] >= 0) & (df['in_spotify_playlists'] <= 2000)]

# 6. Handling Encoding and Formatting Issues
# Example: Remove special characters from song names
df['track_name'] = df['track_name'].str.replace('[^a-zA-Z0-9\s]', '')

# 7. Validating Data
# Example: Check if popularity scores are within a valid range
df['valid_popularity'] = (df['in_spotify_playlists'] >= 0) & (df['in_spotify_playlists'] <= 2000)

# 8. Removing Irrelevant Columns
# Example: Drop columns that are not needed for analysis
df.drop(['bpm', 'mode'], axis=1, inplace=True)

# 9. Data Profiling
# Example: Display summary statistics and data profile
data_profile = df.describe()

# 10. Testing Data Quality
# Example: Perform data quality checks
data_quality_check = df['valid_popularity'].value_counts()

# Display the cleaned dataset
print(df.head())
print(df)

# Display data profile and data quality check results
print(data_profile)
print(data_quality_check)
```
